```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
dataset=[['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
         ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
         ['Milk',],
         ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
         ['Corn','Onion','Onion','Kidney Beans','Ice cream','Eggs']
         ]
```

```python
dataset
```

```python
freq_itemsets=pd.DataFrame(dataset)
freq_itemsets
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
frequent_itemsets=apriori(df,min_support=0.6,use_colnames=True)
frequent_itemsets
```

```python
#build association rule using support metric
rules=association_rules(frequent_itemsets,metric='support',support_only=True,min_threshold=0.1)
#use only support metric for Eclat algorithm
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support']]
rules
```

```python
import pyfpgrowth
#creating sample transaction
```

```python
transactions=[['Milk','Bread','Saffron'],
         ['Milk','Saffron'],
         ['Bread','Saffron','Wafer'],
         ['Bread','Saffron','Wafer'],
         
         ]
```

```python
transactions
```

```python
#finding the frequent pattern with min_support=0.5
FrequentPatterns=pyfpgrowth.find_frequent_patterns(transactions=transactions,support_threshold=0.5)
FrequentPatterns
```

```python
#Generating rules with min confidence threshold=0.5
Rules=pyfpgrowth.generate_association_rules(patterns=FrequentPatterns,confidence_threshold=0.5)
Rules
```
